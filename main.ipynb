```python
import subprocess

channel_url = "https://www.youtube.com/@centralchurch5467/videos"
```

```python
subprocess.run(f'yt-dlp -i --print "%(webpage_url)s" "{channel_url}" > urls.txt', shell=True)
```

```python
# yt-dlp 명령어 실행
subprocess.run(f'yt-dlp --flat-playlist -i --print "%(webpage_url)s" "{channel_url}" > urls.txt', shell=True)
```

```python
# urls.txt에 저장된 모든 유튜브 영상을 mp3로 변환하여 downloads 폴더에 저장
subprocess.run('yt-dlp -f bestaudio --extract-audio --audio-format mp3 --audio-quality 192K -o "downloads/%(title)s.%(ext)s" -a urls.txt', shell=True)
```
